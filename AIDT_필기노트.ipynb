```python
result = k.find('this')	      # index of first instance of string; return index
result #this가 두 개일 때는 첫번째가 나옴
```

```python
j = k.strip()      	          # a copy of s without leading or trailing whitespace
j #공백을 제거해라, 부호를 제거해라! .strip
```

```python
tokens = j.split(' ')	        # split s into a list  (whitespace by default) 공백을 기준으로 잘라라! 띄어쓰기 기준으로 자르라는 뜻
tokens #전체 문자열을 토큰화함
```

```python
m = ' '.join(tokens)	        # combine the words of the text into a string using m as the glue
m #자른 것을 합쳐라
```

```python
v = m.replace('this', 'that') # replace instances of t with u inside m
v #바꿔라! 하지만 위험해서 잘 안 함
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

# Define a list of auxiliary verbs (조동사)
auxiliary_verbs = ['be', 'am', 'is', 'are', 'was', 'were', 'being', 'been',
                   'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing',
                   'shall', 'should', 'will', 'would', 'may', 'might', 'must',
                   'can', 'could']

# Get the default English stop words
stop_words = set(stopwords.words('english'))

# Add the auxiliary verbs to the stop words set
stop_words.update(auxiliary_verbs)

# Now use the updated stop_words in your list comprehension
no_stop_gutenberg = [w for w in words if not w in stop_words]
```

```python
fd_nostop = nltk.FreqDist(no_stop_gutenberg).most_common(20)
fd_nostop
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to change the font colors
def gradient_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    shades_of_blue = ["#c41454", "#32a879", "#32a879", "#c2c73a", "#351182"]
    return shades_of_blue[len(word) % len(shades_of_blue)]


# Generate word cloud
wordcloud = WordCloud(width = 800, height = 600,
                      background_color ='white',
                      min_font_size = 10,
                      color_func=gradient_color_func  # Use the color function
                     ).generate(text)

# Display the word cloud using matplotlib
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 2)  # margin space

plt.show()
```
